```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student_performance.csv")

print("Dataset shape (rows, columns):", df.shape)
print("Column name:", df.columns.tolist())

print("First 5 rows of the dataset:")
print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
print(df.info())
```

```python
print("Average score by columns:")

print(df.mean(numeric_only=True))
```

```python
print("Average score by columns:")

print(df.max(numeric_only=True))
```

```python
print("Average score by columns:")

print(df.min(numeric_only=True))
```

```python
# Correlation
print("Correlation:")

print(df.corr(numeric_only=True))
```

```python
df["total score"] = df[["class_participation", "weekly_self_study_hours", "attendance_percentage"]].sum(axis=1)
df["average score"] =df["total score"] / 3
df["passed"] = np.where(df["average score"] >= 50, "Yes", "No")

print("New columns preview (total, average, passed):")
print(df[["total score",  "average score", "passed"]].head())
```

```python
test_prep_group = df.groupby('student_id')[["class_participation", "weekly_self_study_hours", "attendance_percentage"]].mean()
print("student score by test preparation course:")
print(test_prep_group)

# Class-based average
gender_avg = df.groupby("class_participation")[["total_score", "weekly_self_study_hours", "attendance_percentage"]].mean()
print("Average score by gender:")
print(gender_avg)
```

```python
# Convert scores to Numpy array
scores = df[["total_score", "weekly_self_study_hours", "attendance_percentage"]].values

# Mean, Std Deviation
print("Numpy Mean:", np.mean(scores, axis=0))
print("Numpy Std:", np.std(scores, axis=0))

# Z-score (normalize)
z_score = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)
print("Z-scores (normalized numbers)")
print(z_score[:5])
# Average 0
# Negative -
# Positive +
```

```python
# How many students scored above 90 average ?
top_scorers = df[df["average score"] >=90]
print("Top scored student above 90:", len(top_scorers))

# How many failed?
failed = df[df["passed"] == "No"]
print("Failed from exam:", len(failed))
```
